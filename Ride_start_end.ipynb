```python
import pandas as pd
```

```python
# Read in output file
df = pd.read_csv("Resources/output.csv")
df.head()
```

```python
# select the columns of start stations
df2 = df.iloc[:,3:7]
df2.head()
```

```python
df3 = df2.groupby(['start station id', 'start station name', 'start station latitude', 'start station longitude'])["start station id"].count().reset_index(name='count').sort_values(['count'], ascending=False)
```

```python
# build start station table. Count only the top 10 most popular "start" stations
df_top_start = df3.head(10)
df_top_start_copy = df_top_start.copy()
df_top_start_copy["start_end"] = "top_start"
```

```python
df_bottom_start = df3.tail(10)
df_bottom_start_copy = df_bottom_start.copy()
df_bottom_start_copy['start_end'] = 'Bottom_start'
```

```python
df_top_start_copy.columns = ["station id", "station name", "latitude", "longitude","start_end","count"]
df_bottom_start_copy.columns = ["station id", "station name", "latitude", "longitude","start_end","count"]
df_top_start_copy
```

```python
df_bottom_start_copy
```

```python
# do the same for the "end" stations
df4 = df.iloc[:,7:11]
```

```python
df5 = df4.groupby(['end station id', 'end station name', 'end station latitude', 'end station longitude'])["end station id"].count().reset_index(name='count').sort_values(['count'], ascending=False)
```

```python
df_top_end = df5.head(10)
df_top_end_copy = df_top_end.copy()
df_top_end_copy['start_end'] = 'Top_end'
df_bottom_end = df5.tail(10)
df_bottom_end_copy = df_bottom_end.copy()
df_bottom_end_copy['start_end'] = 'Bottom_end'
df_top_end_copy
```

```python
df_top_end_copy.columns = ["station id", "station name", "latitude", "longitude","start_end","count"]
df_bottom_end_copy.columns = ["station id", "station name", "latitude", "longitude","start_end","count"]
df_top_end_copy
```

```python
df_joined = pd.concat([df_top_start_copy,df_top_end_copy,df_bottom_start_copy,df_bottom_end_copy])
df_joined
```

```python
df_joined.to_csv("Resources/start_end.csv",index=False)
```
