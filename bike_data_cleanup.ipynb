```python
import pandas as pd
import glob
#pip install geopy
from geopy import distance
```

```python
path = r'D:\Documents\James\USC Data Analytics Boot camp\USC Data Analytics\ClassRepo\unit_20_tableau\homework\Instructions'
```

```python
all_files = glob.glob(path + "/*.csv")
```

```python
# read in all csv files sequentially
li = []
for file_name in all_files:
    df = pd.read_csv(file_name, index_col=None, header=0)
    li.append(df)
```

```python
# append them all to a single dataframe
frame = pd.concat(li,axis=0, ignore_index=True)
```

```python
# compute age
frame['Age'] = 2021 - frame["birth year"].astype(int)
```

```python
frame.head()
```

```python
# compute the distance between starting and ending points
frame["start coord"] = list(zip(frame["start station latitude"], frame["start station longitude"]))
frame["end coord"] = list(zip(frame["end station latitude"], frame["end station longitude"]))
```

```python
dist=[]
for i in range(len(frame)):
    dist.append(distance.distance(frame.iloc[i]["start coord"],frame.iloc[i]["end coord"]).miles)
```

```python
frame['distance'] = dist
```
